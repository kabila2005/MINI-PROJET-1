# **SUBSET SELECTION PROBLEM**

```python
#Find all the subsets from a set of numbers whose sum is zero.
#Constraint: Subset size must be 5
#Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
```

```python
import random

def find_zero_sum_subsets_random(nums, subset_size, iterations):
    result = set()

    for _ in range(iterations):
        subset = tuple(random.sample(nums, subset_size))
        if sum(subset) == 0:
            result.add(subset)

    return list(result)
numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
iterations = 10000
subsets = find_zero_sum_subsets_random(numbers, 5, iterations)

subsets
```

```python
#Find all the subsets from a set of numbers whose sum is zero.
#Constraint: Subset size must be 3 to 6 only
#Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
```

```python
import random

def find_zero_sum_subsets_random(nums, min_size, max_size, iterations):
    result = set()

    for _ in range(iterations):
        subset_size = random.randint(min_size, max_size)
        subset = tuple(random.sample(nums, subset_size))
        if sum(subset) == 0:
            result.add(subset)

    return list(result)
numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

iterations = 10000
subsets_random = find_zero_sum_subsets_random(numbers, 3, 6, iterations)

subsets_random
```
